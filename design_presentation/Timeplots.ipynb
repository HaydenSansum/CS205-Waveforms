```python
songs = [1,2,5,10,20,40]
converting = np.array([1.176, 2.764, 7.327, 12.8021,23.442, 52.569])
digitizing = np.array([1.819, 3.099, 6.731, 13.093, 23.726, 52.011])
scraping = np.array([1.73700, 2.92533333, 7.56333333, 13.3475, 43.71825, 70.816])
```

```python
plt.figure(figsize=(10,8))
plt.plot(songs,scraping,'-o',label='Scraping from FMA website')
plt.plot(songs,converting,'-o',label='Converting from mp3 to wav')
plt.plot(songs,digitizing,'-o',label='Digitizing (binning)')
plt.xticks(songs)
plt.xlabel('Number of Songs')
plt.ylabel('Runtime (s)')
plt.title('Runtime time for various song processing steps')
plt.legend()
plt.show()
```
